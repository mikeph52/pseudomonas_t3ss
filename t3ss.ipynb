# Type III Secretion system gene analysis on _Pseudomonas syringae_ pathovars

## Introduction
_Pseudomonas syringae pv. tomato DC3000_ gains its virulence factor from the **type III secretion system** effectors, notably the **AvrE1**
and the **HopM1**. In this experiment, we aim to isolate effector-encoding genes from a genome assembly of _Pseudomonas syringae pv. syringae_. A major limitation of this approach is the absence of a well-annotated reference genome for this pathovar, which complicates accurate gene identification and functional annotation.

## Step 1: translate 

```python
from Bio import SeqIO

# data absolute paths
df_pst = "data/GCF_000007805.1_ASM780v1_genomic.fna"
df_pss = "data/GCF_043515225.1_ASM4351522v1_genomic.fna"

# ret
pst = list(SeqIO.parse(df_pst, "fasta"))
pss = list(SeqIO.parse(df_pss, "fasta"))

for record in pst:
    print(record.id)
    print(record.description)
    print(record.seq)
    print(len(record.seq))

for record in pss:
    print(record.id)
    print(record.description)
    print(record.seq)
    print(len(record.seq))
```

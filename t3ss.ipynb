# Type III Secretion system gene analysis on _Pseudomonas syringae_ pathovars

## Introduction
_Pseudomonas syringae pv. tomato DC3000_ gains its virulence factor from the **type III secretion system** effectors, notably the **AvrE1**
and the **HopM1**. In this experiment, we aim to isolate effector-encoding genes from a genome assembly of _Pseudomonas syringae pv. syringae_. A major limitation of this approach is the absence of a well-annotated reference genome for this pathovar, which complicates accurate gene identification and functional annotation.

## Step 1: translate 

```python
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord

# data absolute paths
df_pst = "data/chromosome_1_NC_004578_1_Pseudomonas_syringae_pv__tomato_str__DC3000__complete_sequence.fasta"
df_pss = "data/GCF_043515225.1_ASM4351522v1_genomic.fna"
df_hopm1 = "data/hopm1.fasta"
df_avre1 = "data/avre1.fasta"

# ret
pst = list(SeqIO.parse(df_pst, "fasta"))
pss = list(SeqIO.parse(df_pss, "fasta"))
hopm1 = list(SeqIO.parse(df_hopm1, "fasta"))
avre1 = list(SeqIO.parse(df_avre1, "fasta"))

# Data and details
print("Type III Secretion system gene analysis on Pseudomonas syringae pathovars")
print("By Mike Philippakis, 2026")
print("\nFor Pseudomonas syringae pv. tomato DC3000: ")
for record1 in pst:
    print(record1.id)
    print(record1.description)
    print(len(record1.seq))

print("\nFor Pseudomonas syringae pv. syringae: ")
for record2 in pss:
    print(record2.id)
    print(record2.description)
    print(len(record2.seq))

print("\nFor HopM1: ")
for record3 in hopm1:
    print(record3.id)
    print(record3.description)
    print(len(record3.seq))

print("\nFor AvrE1:")
for record4 in avre1:
    print(record4.id)
    print(record4.description)
    print(len(record4.seq))

prot_pst = [SeqRecord(record1.seq.translate(table=1, to_stop=False, cds=False), id=record1.id, description="Pseudomonas syringae pv. tomato DC3000 translated")]
prot_pss = [SeqRecord(record2.seq.translate(table=1, to_stop=False), id=record2.id, description="Pseudomonas syringae pv. syringae translated")]

SeqIO.write(prot_pst, "data/pst_translated.fasta", "fasta")
SeqIO.write(prot_pss, "data/pss_translated.fasta", "fasta")
```
